```python
x = bc['data']
y = bc['target']
```

```python
x_train,x_test,y_train,y_test = sklearn.model_selection.train_test_split(x,y,test_size=.4)
```

```python
reg = sklearn.linear_model.LogisticRegression()
```

```python
reg.fit(x_train,y_train)
reg.score(x_test,y_test)
```

```python
clf_gini = sklearn.tree.DecisionTreeClassifier(criterion="gini",random_state=100,max_depth=3,min_samples_leaf = 5)
```

```python
clf_gini.fit(x_train,y_train)
clf_gini.score(x_test,y_test)
```

```python
poly = sklearn.preprocessing.PolynomialFeatures(degree = 2)
x_poly = poly.fit_transform(x)
x_train,x_test,y_train,y_test = sklearn.model_selection.train_test_split(x_poly,y,test_size=.25)
reg1= sklearn.linear_model.LinearRegression()
reg1.fit(x_train,y_train)
reg1.score(x_test,y_test)
```

```python
clf1 = sklearn.ensemble.RandomForestRegressor(random_state=0)
```

```python
clf1.fit(x_train,y_train)
clf1.score(x_test,y_test)
```
