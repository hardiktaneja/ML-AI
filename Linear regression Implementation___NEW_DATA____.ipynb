```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
labeldata = [(2*x +1) for x in range(0,10)]
```

```python
labeldata
```

```python

def createfxlist(b0,b1):
    fxlist=[]
    for i in range(0,10):
        fxlist.append(b0 + b1*i)
    return fxlist
```

```python
# Change in Cost Function
#w.r.t b0
#w.r.t b1
# dcb0 = 2/len(traindata) * (labeldataarray - fxarray)
# dcb1 = 2/len(traindata) * (labeldataarray - fxarray)
def updatevals(fxlist,b0,b1):
    dcb0 = 0
    cost=0
    dcb1= 0
    for i in range(len(fxlist)):
        dcb0 +=-1 *2/len(fxlist) * (labeldata[i] - fxlist[i])
        dcb1 += -1*2/len(fxlist) * i*(labeldata[i] - fxlist[i])
        cost += 1/len(fxlist) * (labeldata[i] - fxlist[i])**2
    alpha = -0.005
    b0 = b0+ alpha*dcb0
    b1 = b1+ alpha*dcb1
    return b0,b1,cost
```

```python
epochs=1400
```

```python
b0list=[]
b1list=[]
costlist=[]
for i in range(epochs):
    if(i==0):
        b0=0
        b1=0
    fxlist = createfxlist(b0,b1)
    b0,b1,cost = updatevals(fxlist,b0,b1)
    b0list.append(b0)
    b1list.append(b1)
    costlist.append(cost)
print("For",i+1,"Epoch :","\nCost",cost,"\nb0 :",b0,"\nb1 :",b1)
```

```python
import matplotlib.pyplot as plt
```
