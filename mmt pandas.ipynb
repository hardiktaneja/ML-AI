```python
def rating(string):
    try:
#         print("intstring")
        return int(string)
    except:
#         print("intstringsplit")
        try:
            return int(string.split(" ")[0])
        except:
            if(string.split(" ")[0] == 'One' or string.split(" ")[0] == 'one'):
#                 print("shit")
                return 1
            if(string.split(" ")[0] == 'Two' or string.split(" ")[0] == 'two'):
                return 2
            if(string.split(" ")[0] == 'Three' or string.split(" ")[0] == 'three'):
                return 3
            if(string.split(" ")[0] == 'Four' or string.split(" ")[0] == 'four'):
                return 4
            if(string.split(" ")[0] == 'Five' or  string.split(" ")[0] =='five'):
                return 5
```

```python
m={}
for i in range(len(x)):
    rate =rating((x['hotel_star_rating'])[i])
    if(rate == 4):
        city=x['city'][i]
        if(city not in m.keys()):
            m[city] = 1
        else:
            m[city]=m[city]+1
```
